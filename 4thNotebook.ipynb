##### Does quality affect price differently based on house size?

## 1} Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2} Load

```python
# Load The DataSet

df = pd.read_csv("/content/features_DataSet.csv")

df.head()
```

## 3} Create size groups

```python
# Create size categories

df["size_group"] = pd.qcut(df["Gr Liv Area"], q=3, labels=["Small", "Medium", "Large"])
```

## 4} Boxplot

```python
# Compare quality effect across size groups

sns.boxplot(x=df["Overall Qual"], y=df["SalePrice"], hue=df["size_group"])
plt.title("Quality vs Price across Size Groups")
plt.show()
```

## Insight

> - quality increases price in all sizes  
> - large houses benefit more from high quality  
> - interaction between size and quality affects price
