# 💻Dashboard

## 1} Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2} Load

```python
# Load The DataSet

df = pd.read_csv("/content/features_DataSet.csv")

df.head()
```

## The dashboard

```python
plt.figure(figsize=(12, 10))

# 1
plt.subplot(3, 2, 1)
plt.hist(df["SalePrice"], bins=30)
plt.title("SalePrice Distribution")

# 2
plt.subplot(3, 2, 2)
plt.hist(df["Overall Qual"], bins=10)
plt.title("Quality Distribution")

# 3
plt.subplot(3, 2, 3)
plt.scatter(df["Gr Liv Area"], df["SalePrice"])
plt.title("Area vs Price")

# 4
plt.subplot(3, 2, 4)
plt.boxplot([df[df["Full Bath"] == i]["SalePrice"] for i in df["Full Bath"].unique()])
plt.title("Price by Bathrooms")

# 5
plt.subplot(3, 2, 5)
plt.hist(df["SalePrice_log"], bins=30)
plt.title("Log SalePrice")

plt.tight_layout()
plt.show()
```
