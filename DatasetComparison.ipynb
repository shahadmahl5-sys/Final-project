## 1} Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2} Load

```python
df2 = pd.read_csv("/content/insurance.csv")

df2.head()
```

## 3} Distribution

```python
# Distribution of medical cost

sns.histplot(df2["charges"], kde=True)
plt.title("Medical Cost Distribution")
plt.show()
```

## 4} Relationship

```python
# Relationship between BMI and charges

sns.scatterplot(x=df2["bmi"], y=df2["charges"])
plt.title("BMI vs Charges")
plt.show()
```

#### Insight

> - charges increase with BMI  
> - higher BMI leads to higher medical cost  
> - distribution is slightly skewed  

## 5} Comparison

> - In the housing dataset, SalePrice depends on quality and living area  
> - In the medical dataset, cost depends on BMI
