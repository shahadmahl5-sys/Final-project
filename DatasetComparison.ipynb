## 1} Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2} Load

```python
df2 = pd.read_csv("/content/final_data_insurance.csv")

df2.head()
```

## 3} Distribution

```python
# Distribution of medical cost

sns.histplot(df2["charges"], kde=True)
plt.title("Medical Cost Distribution")
plt.show()
```

## 4} Relationship

```python
# Relationship between BMI and charges

sns.scatterplot(x=df2["bmi"], y=df2["charges"])
plt.title("BMI vs Charges")
plt.show()
```

#### Insight

> - Medical charges are right-skewed, most values are low with some high outliers  
> - Charges tend to increase as BMI increases  
> - Higher BMI is associated with higher medical costs
