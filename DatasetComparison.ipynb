## 1} Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2} Load

```python
df2 = pd.read_csv("/content/final_data_insurance.csv")
# This is the cleaned dataset by Tala Fahed

df2.head()
```

```python
df = pd.read_csv("/content/features_DataSet.csv")
# This is My cleaned dataset

df.head()
```

## 3} Distribution

```python
# Distribution of medical cost

sns.histplot(df2["charges"], kde=True)
plt.title("Medical Cost Distribution")
plt.show()
```

```python
# Distribution of house prices

sns.histplot(df["SalePrice"], kde=True)
plt.title("SalePrice Distribution")
plt.show()
```

## 4} Relationship

```python
# Relationship between BMI and charges

sns.scatterplot(x=df2["bmi"], y=df2["charges"])
plt.title("BMI vs Charges")
plt.show()
```

```python
# Relationship between area and price

sns.scatterplot(x = df["Gr Liv Area"], y = df["SalePrice"])
plt.title("Area vs Price")
plt.show()
```

## Dataset Comparison

##### - Both datasets show a positive relationship between features and the target
##### - Increasing key features increases the target value in both datasets
##### - Clear relationships appear in the visualizations (distribution and scatter plots)

##### - Differences:
#####   - Housing dataset depends on property features (area, quality)
#####   - Medical dataset depends on health features (BMI)

##### - Conclusion:
#####   - Housing dataset is more complex (many factors)
#####   - Medical dataset is more direct and simpler
