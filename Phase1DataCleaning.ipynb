## 7} Column lists

```python
# Drop column has more than 80% missing valus
drop_cols = ["Pool QC", "Misc Feature", "Alley", "Fence"]

# Feature doesn't exist in this house
cat_cols = [
    "Mas Vnr Type", "Fireplace Qu",
    "Bsmt Qual", "Bsmt Cond", "Bsmt Exposure", "BsmtFin Type 1", "BsmtFin Type 2",
    "Garage Type", "Garage Finish", "Garage Qual", "Garage Cond"
]
```

```python
# Skewed so MEDIAN is safer than MEAN
median_cols = [
    "Lot Frontage","Garage Yr Blt",
    "BsmtFin SF 1", "BsmtFin SF 2",
    "Bsmt Unf SF", "Total Bsmt SF",
    "Garage Cars", "Garage Area"
]

# Fill with ZERO
zero_fill_cols = ["Bsmt Full Bath", "Bsmt Half Bath", "Mas Vnr Area"]
```

## 8} fix_dtypes

```python
def fix_dtypes(df):
    out = df.copy()

    # Convert categorical numeric columns to string
    out['MS SubClass'] = out['MS SubClass'].astype(str)

    # Garage Yr Blt should be numeric
    out['Garage Yr Blt'] = pd.to_numeric(out['Garage Yr Blt'], errors='coerce')

    return out
```

## 9} fill_missing

```python
def fill_missing(df):

    out = df.copy()

    # Drop columns with >80% missing
    out = out.drop(columns=drop_cols)

    # # Fill categorical missing with 'None' (feature does not exist)
    for col in cat_cols:
            if col in out.columns:
                out[col] = out[col].fillna("None")

    # Fill numeric skewed with MEDIAN
    for col in median_cols:
            if col in out.columns:
                out[col] = out[col].fillna(out[col].median())

    # Fill with ZERO
    for col in zero_fill_cols:
        if col in out.columns:
            out[col] = out[col].fillna(0)

    # Fill Electrical column with MODE
    out["Electrical"] = out["Electrical"].fillna(out["Electrical"].mode()[0])

    return out
```

#### Handling Missing Values


> - Drop columns with more than 80% missing values because they won't help us.
>
> - For categorical columns where NaN indicates the feature does not exist, fill missing values with "None".
> - For numeric columns with extreme outliers, use median to fill missing values.
> - For numeric columns without extreme outliers (symmetric), use mean to fill missing values.
> - For numeric count columns where NaN indicates the feature does not exist, fill with 0.

## 10} remove_duplicates

```python
def remove_duplicates(df):

    out = df.copy()

    out = out.drop_duplicates()
    # Any duplicates will drop

    return out
```

## 11} cap_outliers

```python
def cap_outliers(df):

    out = df.copy()

    # Calculate IQR
    Q1  = out["SalePrice"].quantile(0.25)
    Q3  = out["SalePrice"].quantile(0.75)
    IQR = Q3 - Q1

    # Define acceptable range
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Print number of outliers before capping
    print("Outliers before capping:", ((out["SalePrice"] < lower) | (out["SalePrice"] > upper)).sum())

    # Cap the values outside the range
    out["SalePrice"] = out["SalePrice"].clip(lower=lower, upper=upper)

    # Print number of outliers after capping
    print("Outliers after capping:", ((out["SalePrice"] < lower) | (out["SalePrice"] > upper)).sum())

    return out
```

## 12} validate

```python
def validate(df):

    # No nulls in key columns
    key_cols = ["SalePrice", "Lot Frontage", "Mas Vnr Area"]

    for col in key_cols:
        assert df[col].isna().sum() == 0, f"Column {col} has missing values"

    # All target values > 0
    assert df["SalePrice"].min() > 0, "SalePrice has non positive values"

    # Dataframe is not empty
    assert len(df) > 0, "DataFrame is empty"

    print("\nData cleaning done successfully❤️")
```

## 13} clean_data

```python
def clean_data(df):

    out = fix_dtypes(df)
    out = fill_missing(out)
    out = remove_duplicates(out)
    out = cap_outliers(out)
    validate(out)

    # Apply all pipline in one function
    return out
```

## 14} Apply

```python
df = clean_data(df)

# See the new shape after cleaning
print("\nshape after:", df.shape)
```
